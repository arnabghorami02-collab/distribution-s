1.Simulate 30 rolls with =RANDBETWEEN(1,6). What is the probability of rolling a 3 exactly 5 times? (Hint: Use
BINOM.DIST)

Sol :      
Trials (n): 30

Probability of success (p): 1/6 (since there is one '3' on a six-sided die)

Successes (x): 5

Excel Formula: =BINOM.DIST(5, 30, 1/6, FALSE)

4.What is the concept of standardization (z-score), and why is it important in data analysis? Explain the
formula and how standardization transforms a dataset.

Sol :       

Concept: Standardization converts data from different scales into a common scale with a mean of 0 and a standard deviation of 1.

Formula: z = \frac{x - \mu}{\sigma}

Importance: It allows you to compare values from different datasets (e.g., comparing an SAT score to an ACT score).

Transformation: It shifts the distribution (centering it at 0) and scales it (compressing or expanding it) without changing the relative shape of the distribution.

5.What is Kurtosis and their type?

Sol :     
Kurtosis measures the "tailedness" of a distribution (how much of the data is in the outliers vs. the peak).

​Mesokurtic: Zero excess kurtosis (Normal Distribution).

​Leptokurtic: Positive excess kurtosis. Features a sharp peak and heavy/fat tails (more outliers).

​Platykurtic: Negative excess kurtosis. Features a flat peak and thin tails (fewer outliers).

6.Explain why the uniform distribution is a good model for the outcome of rolling a fair die.

Sol :        
A fair die is a classic example of a Discrete Uniform Distribution because:

- There is a finite number of outcomes (1, 2, 3, 4, 5, 6).
- Each outcome has an equal probability (1/6) of occurring.
- No single number is favored over another, resulting in a "flat" probability mass function.

7.Use Excel to compute the probability of getting at least 8 successes in 15 trials with success probability 0.5

Sol :     
To find "at least 8," you calculate 1 minus the probability of getting 7 or fewer.

Trials: 15

Probability: 0.5

Excel Formula: =1 - BINOM.DIST(7, 15, 0.5, TRUE)

8.How does log transformation help in stabilizing variance and making data more normally distributed?

Sol :     
- Stabilizing Variance: Log transformations compress the higher end of a scale more than the lower end. This helps manage "heteroscedasticity," where the spread of data increases as values get larger.

- Normality: It is highly effective for right-skewed data (data with a long tail on the right, like income or population). By pulling the extreme outliers inward, it makes the distribution more bell-shaped and symmetrical.
